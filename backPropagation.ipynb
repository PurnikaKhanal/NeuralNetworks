```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'Math_Marks':[94,80,70,58,50],
                 'Physics_Marks':[85,75,67,49,45],
                 'Attendance':[95,84,75,60,55],
})
```

```python
parameters={'w1':np.random.randn(2, 2) * 0.01,
            'b1':np.zeros((2, 1)),
            'w2':np.random.randn(2, 1) * 0.01,
            'b2':np.zeros((1, 1))
            }
parameters
```

```python
vals=df.to_numpy()
vals
```

```python
A=np.delete(vals, 2, axis=1)
A
```

```python
def forward_to_layer(A, parameters):
    Z1 = np.dot(A, parameters['w1']) + parameters['b1'].T
    A_new = 1/(1+np.exp(-Z1))
    return A_new

A_layer = forward_to_layer(A, parameters)
A_layer
```

```python
def forward_to_output(A_new, parameters):
    Z2 = np.dot(A_new, parameters['w2']) + parameters['b2'].T
    A_output = 1/(1+np.exp(-Z2))
    return A_output

A_output=forward_to_output(A_layer, parameters)
A_output
```

```python
test=np.ones((A.shape[0], 1)).T
test
```

```python
def backward_pass(A, A_layer, A_output, y, parameters, learning_rate):
    dA_output = A_output - y

    dw1 = np.dot(A.T, dA_layer)
    dw2 = np.dot(A_layer.T, dA_output)

    db2 = np.dot(np.ones((1, A.shape[0])), dA_output)
    db1 = np.dot(np.ones((1, A.shape[0])), dA_layer)

    dA_layer = np.dot(dA_output, parameters['w2'].T)

    parameters['w1'] = parameters['w1'] - learning_rate * dw1
    parameters['b1'] = parameters['b1'] - learning_rate * db1
    parameters['w2'] = parameters['w2'] - learning_rate * dw2
    parameters['b2'] = parameters['b2'] - learning_rate * db2

    return parameters

y = vals[:, 2:3]
y_min, y_max = y.min(), y.max()
y = (y - y_min) / (y_max - y_min)

A = np.delete(vals, 2, axis=1)
A_min, A_max = A.min(axis=0), A.max(axis=0)
A = (A - A_min) / (A_max - A_min)
```

```python
def error(y, A_output):
    return np.mean((y - A_output)**2)
```

```python
epochs=2000
for i in range(epochs):
    A_x=forward_to_layer(A, parameters)
    A_output=forward_to_output(A_x, parameters)
    parameters=backward_pass(A, A_x, A_output, y, parameters, 0.1)
    err = error(y, A_output)
    if (i+1) % 100 == 0:
        print(f"Epoch {i+1}, Error: {err:.6f}")
parameters
```

```python
def test(Maths, Physics):
    input_data = np.array([[Maths, Physics]])
    input_data = (input_data - A_min) / (A_max - A_min)
    A_new=forward_to_layer(input_data, parameters)
    A_output=forward_to_output(A_new, parameters)
    predicted_attendance = A_output * (y_max - y_min) + y_min
    print("Predicted Attendance:")
    print(predicted_attendance)
test(85, 90)
```
